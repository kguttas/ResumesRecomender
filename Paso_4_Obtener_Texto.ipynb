```python
import os
from os import listdir
from os.path import isfile, join
import shutil

import pytesseract
import cv2

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from Database import db_mongo
```

```python
path_temp_folder_images = "data/resumes/temp/"
```

```python
path_directories = [x[0]  for x in os.walk(path_temp_folder_images)]

#random.shuffle(path_directories)

folder_rectagles = list(filter(lambda x: '_rectangles' in x and 'cuts_' in x, path_directories))

print(folder_rectagles)
```

```python
def get_files_in_directory(path_directory, extension):

    onlyfiles = [f for f in listdir(path_directory) if isfile(join(path_directory, f))]

    files_names = list(filter(lambda x: '.' + extension in x, onlyfiles))
    
    files_names.sort()
    
    return files_names
```

```python
def get_text_from_image(path_image):
    
    image = cv2.imread(path_image, 0)
    
    #fig = plt.figure()
    #fig.set_size_inches(10,23)
    #plt.title('image')
    #plt.imshow(image, cmap='gray', aspect='auto') 
    #plt.show()
  
    #gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

    blur = cv2.GaussianBlur(thresh, (3, 3), 0)
    
    #plt.title('blur')
    #plt.imshow(blur, cmap='gray', aspect='auto') 
    #plt.show()

    data = pytesseract.image_to_string(blur, lang='eng', config='--oem 3 --psm 6')
    
    #df = pytesseract.image_to_data(blur, output_type='data.frame')
    #df = df[df['conf'] != -1] 
    #df['text'] = df['text'].apply(lambda x: x.strip())
    #df = df[df['text']!=""]
    #df['text'] = df['text'].apply(lambda x: x.upper())
    
    return data
```

```python

db = db_mongo.db_mongo()

for directoriesRectangles in folder_rectagles:
    
   
    print(directoriesRectangles)
    
    parts_path_folder = directoriesRectangles.split('/')
    
    last_part_with_class = [x for x in parts_path_folder if 'cuts_' in x]
    
    CLASS = 'NONE'
    
    if len(last_part_with_class) > 0:
        parts_last_part = last_part_with_class[0].split('_')
    
        CLASS = parts_last_part[2]
    
    print(CLASS)
    
    cuts_files_jpeg = get_files_in_directory(directoriesRectangles, 'jpeg')
    
    for name_cut_image in cuts_files_jpeg:
        
        print(name_cut_image)

        path_cut_image = join(directoriesRectangles, name_cut_image)
        text_from_image = get_text_from_image(path_cut_image)

        #print(text_from_image)
        
        db.insert_text({"text": text_from_image, "class": CLASS})
        
        #text_from_image.head()
    
```
