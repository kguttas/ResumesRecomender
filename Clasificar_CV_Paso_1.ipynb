```python
db = db_mongo.db_mongo()

path_directory_resumes = "data/resumes/estandar"
```

```python
def get_files_in_directory(path_directory, extension):

    onlyfiles = [x  for x in os.walk(path_directory)] #[f for f in os.walk(path_directory) if isfile(join(path_directory, f))]

    #print(onlyfiles[1])
    
    list_files = list()
    
    for folder in onlyfiles:
        
        path_base_folder_pdf_0 = folder[0]
        
        path_base_folder_pdf_1 = folder[1]
        
        path_base_folder_pdf_2 = folder[2]
        
        #print(path_base_folder_pdf_0)
        
        #print(path_base_folder_pdf_1)
        
        #print(path_base_folder_pdf_2)
        
        
        
        for file_name in path_base_folder_pdf_2:
            
            
            
            path_file_resume = path_base_folder_pdf_0 + "/" + file_name
            #print(path_file_resume)
            list_files.append(path_file_resume)
        
    
    files_names = list(filter(lambda x: '.' + extension in x, list_files))
    
    files_names.sort()
    
    return files_names
```

```python
all_paths_files = get_files_in_directory(path_directory_resumes, 'pdf')
```
